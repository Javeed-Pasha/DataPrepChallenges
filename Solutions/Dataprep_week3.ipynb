```python
# Read an excel File , use sheet_name=None to read all excel sheets
all_dfs = pd.read_excel(r'C:\Users\jpasha\Desktop\Data_Prep\Data_2022_w3.xlsx', sheet_name=None)
```

```python
#Add an addional column  called store and save the value corresponding to its sheet name.
sheetlists = list(all_dfs.keys())
for sheet in sheetlists:
    if sheet in all_dfs.keys():
        all_dfs[sheet]['Store']=sheet
```

```python
# verify that a column named Stoer is added with sheet name value 
all_dfs['Leeds']
```

```python
#concaenate all sheets into a single dataframe and reset the index .
df = pd.concat(all_dfs, ignore_index=True)
```

```python
df.head()
```

```python
#Unpivot = df.melt(id_vars = 'Date', var_name = 'Product', value_name = 'Sales')
#id_vars: identifies the column to be used as identifier variables
#value_vars: the columns to unpivot. Which, if empty, (as is the case here), uses all columns except those identified in id_vars
#var_name: assigns a name to the columns that were unpivoted
#value_name: assigns a name to to the value column
 
```

```python
# use df.melt to unpivote columns to rows 
Unpivot = df.melt(id_vars = ['Date','Store'],var_name = 'Product', value_name = 'Sales')  

Unpivot.head()
```

```python
# split the  product name (eg.New - Saddles) into Customertype= New and Product=Saddles 
Unpivot[['CustomerType','Product']] =Unpivot['Product'].str.split("-", n=2, expand = True).loc[:,[0,1]]
```

```python
Unpivot.head()
```

```python
# add a Quarters column deried from date column 
Unpivot['Quarter'] =Unpivot['Date'].dt.quarter
Unpivot.head()
```

```python
# aggregate the products sold per Product and quarter
Output1=Unpivot.groupby(
     ['Product','Quarter']
 ).agg({ # since my pandas verion is <0.25 i couldnt use named aggregations
  #   'Quantity' :[ ('Quantity','sum')],
     'Sales' : [('Sales','sum')],
})
Output1.columns = ['Products Sold'] 
Output1
```

```python
# aggregate the products_sold per Product,CustomerType and Store.
Output2=Unpivot.groupby(
     ['Store','CustomerType','Product']
 ).agg({ # since my pandas verion is <0.25 i couldnt use named aggregations
  #   'Quantity' :[ ('Quantity','sum')],
     'Sales' : [('Sales','sum')],
})
Output2.columns = ['Products Sold'] 
#y.to_csv("sales_by_Brand_BikeType.csv")
Output2
```
