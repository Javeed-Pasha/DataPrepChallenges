```python
Unpivot = df.melt(id_vars = ['Date','Store'],var_name = 'Product', value_name = 'Sales')  
 
```

```python
 
Unpivot[['CustomerType','Product']] =Unpivot['Product'].str.split("-", n=2, expand = True).loc[:,[0,1]]
```

```python
Unpivot.head()
```

```python
Unpivot['Quarter'] =Unpivot['Date'].dt.quarter
Unpivot.head()
```
