#### Learned how to access specific rows/ columns examples

```python
#loc is label based, iloc is index based 
#Syntax iloc/loc[row,column] , Use : to represent All, to specify a range(slicing) use index/Label: index/label
# return a single ROW from dataframe  returns as a SERIES
print( type(df.loc[0])) #returns a dataframe.Series

print(    (df.loc[0,'Date'])) # first row date column 

print(     (df.loc[:,['Date','Order ID']])) #  date and orderid  column  of all rows 
```

```python
 
#if use [] in loc[] returns a dataframe 
print(df.loc[[0]])
# returns ROWS as a dataframe 
print(type(df.loc[[0,1]]))
print(df.loc[[0,1]])

```

```python
#Callable function
# Selecting columns
df.loc[[0,1], lambda df: ['Order ID', 'Date']]

df.iloc[lambda df: [0,1], :]
```

```python
# With condition
df.loc[lambda df: df['Customer Age'] > 70, :]
# To filter data with callable, iloc will require list() to convert the output of conditions into a boolean list
df.iloc[lambda df: list(df['Customer Age'] > 70), :] 
```

####  data cleansing   examples

```python
df.dropna(inplace=True) # inplace modifies the original  dataframe  df
df.fillna(999,inplace=True)  #fills any null value with 999 in any column 
df.dropna(subset=['Date'],inplace=True) # drop any row where date is null 
df['Existing Customer?'].fillna('No',inplace=True) # fills any null values in column "Exiting customer?"
#  use dataframes  mean() median() and mode() methods  to replce missing values rather than with constant values 
df['Date'] = pd.to_datetime(df['Date']) #  if datetime in different formats use dataframe.to_datetime() 

print(df.info()) # helpful in data profiling to identify total records , null % , column datatypes 
```

### Actual data challenge Solution

```python
#1 Red data from excel . 
import pandas as pd
df = pd.read_excel (r'C:\Users\jpasha\Desktop\Data_Prep\Data_2022_w1.xlsx')
# 2. To split the Store-Bike column into Bike and store columns
#    Google searched how to split a column , found out dataframe.Series has a split function
#    from our exploration of  dataframe above dataframe['columnname'] returns a series 
#    chain that output to loc function to extract column 1 and 3
df[['Store','Bike']] =df['Store - Bike'].str.split(" ", n=2, expand = True).loc[:,[0,2]]
# 3. the next task was to split the  date into quarter and   monthNumber 
#   Googled how to get quarter from a datetime in pandas
df['Quarter'] =df['Date'].dt.quarter
df['Day of Month'] =df['Date'].dt.day

# 5. Drop columns
#df.drop('Store - Bike', inplace=True, axis=1)
df.drop(['Store - Bike','Date'], inplace=True, axis=1)

# 4. Re-arrange the columns,not necessary but still explored index function 
df.sort_index(axis=1)

# 5. saved the output to a file , to be consumed by others
df.to_excel("Data_2022_w1_Solution.xlsx",index=False) 

print('The Cleansed data')
df
#This exercise  helped to understand whats pandas dataframe vs Series
# Explored datacleansing functions and Transform data in pandas

 
```
